Первое задание

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5,6,7])
a = np.arange(1,8)
print(x)
print(a)
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7.])
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y,
           c = 'deeppink')    
ax.set_facecolor('grey')     #  цвет области Axes
ax.set_title('диаграмма рассеяния')     #  заголовок для Axes

fig.set_figwidth(8)     #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```

Второе задание

```python

%config InlineBackend.figure_format = 'svg'
t = np.linspace(0, 5, 51)
f = np.cos(t)

```

```python
plt.plot(t,f,c = 'green')
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

label_font = {
    "fontsize": 9,
    "family": "serif",
}

plt.title("График f(t)", fontdict=title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения  f", fontdict=label_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```
